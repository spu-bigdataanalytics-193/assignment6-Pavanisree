```python
# Reading data
file_location = "/FileStore/tables/All_Beauty.json"
file_type = "json"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df1 = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df1)
```

```python
# 3 Tasks
# Creating a table in SQL
df1.createOrReplaceTempView("df1_table")
```

```python
# Task 1 Showing 10 frist columns
display(spark.sql("""
  select * from 
  df1_table
  limit 10"""))
```

```python
spark.sql("SELECT MAX(overall) as maxval FROM df1_table").collect()[0].asDict()['maxval']
```

```python
# Task 2Group by and get max, min, count of a column in the dataset.
display(spark.sql("""
  select max(overall) as maxvalue, min(overall) as minvalue, count(overall) as cnt, reviewerID from df1_table group by reviewerID """))
```

```python
# Task 3 Apply group by with having clause.
display(spark.sql("""
  select  count(overall) as cnt, reviewerID from df1_table group by reviewerID having count(overall) = 1 """))
```

```python
# Task 4 Printing schema 
df = sqlContext.sql("SELECT * FROM df1_table")
df.printSchema()
```

```python
#Task 5 Filter your dataset by some conditions based on your column.
display(spark.sql("""
  select * from 
  df1_table
  where reviewerID = 'A1V6B6TNIC10QE' """))
```

```python
#Task 6 Apply order by.
display(spark.sql("""
  select * from 
  df1_table
  order by reviewerID  """))
```

```python
# Task 7 Apply inner join/ left join/ right join on your two tables.
display(spark.sql("""
  select a.image, a.overall, a.reviewText, a.reviewerID, a.reviewerName, a.style
  from df1_table a
  inner join df1_table b on a.reviewerID = b.reviewerID
  """))
```

```python
# Task 8 Select distinct records by a column.
display(spark.sql("""
  select distinct a.reviewerName
  from df1_table a
  """))
```
